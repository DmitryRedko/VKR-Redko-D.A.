```python
import pandas as pd
from datetime import datetime, timedelta
import locale

locale.setlocale(locale.LC_TIME, 'ru_RU.UTF-8')
```

```python
df = pd.read_csv('./SBER_news_in_progress.csv')
df.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
def format_date(date_str):
    try:
        try:
            return datetime.strptime(date_str, '%d %m %Y').strftime('%Y-%m-%d')
        except:
            try:
                return datetime.strptime(date_str, '%d %B %Y').strftime('%Y-%m-%d')
            except: raise Exception
    except:
        print(date_str)
        raise Exception
    
df['date'] = df['date'].apply(lambda x: format_date(x))
df
```

```python
for i in range(10):
    df['content'] = df['content'].replace('\n\n', '\n', regex=True)
    df['filtered_content'] = df['content'].replace('\n\n', '\n', regex=True)
    df['EN_filtered_content'] = df['content'].replace('\n\n', '\n', regex=True)
```

```python
df.to_csv('SBER_news_filtered.csv', index=False)
```
